```python
def viewPercentiles(data, percentile_interval=5):
    percentiles = np.arange(percentile_interval, 100, percentile_interval) # [5, 10, 15, ..., 95]
    p_values = np.percentile(data, percentiles)

    for p, value in zip(percentiles, p_values):
        print(f"{p}th percentile: {value}")
```

```python
# Correlation between velocity and change in position

for ind in range(len(train_data)):
    del_x = (train_data[ind, 0, :, 0][1:] - train_data[ind, 0, :, 0][:-1]).flatten()
    vel_x = train_data[ind, 0, :, 2][:-1].flatten()

    corr = np.corrcoef(del_x, vel_x)
    if corr[0, 1] < 0.2:
        print(f"Index={ind}, Correlation={corr}")
```

```python
x = train_data[:, 0, :, 1]
del_x = (x[:, 1:] - x[:, :-1]).flatten()
vel_x = train_data[:, 0, :, 3][:, :-1].flatten()

factors = vel_x / del_x
# print(vel_x[:5])
# print(del_x[:5])
# print(x[:5])
print(f"{factors[:5]}")
print(f"Mean factor={np.mean(factors)}")

# Assuming 'arr' is your numpy array
percentiles = np.arange(5, 100, 5) # [5, 10, 15, ..., 95]
p_values = np.percentile(factors, percentiles)

for p, value in zip(percentiles, p_values):
    print(f"{p}th percentile: {value}")
```
